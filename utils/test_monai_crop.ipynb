```python
import monai
from monai.transforms import LoadImaged, EnsureChannelFirstd, Orientationd, Spacingd
from monai.data import DataLoader, Dataset
import os
```

```python

# data_dir = "/path/to/data"
data_dir = '../../ncct_cect/vindr_ds/batches/batch10/1.2.840.113619.2.278.3.717616.202.1592287871.359/1.2.840.113619.2.278.3.717616.202.1592287871.364.3/'
files = [{"image": os.path.join(data_dir, d)} for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))]
```

```python

transforms = monai.transforms.Compose([
    LoadImaged(keys=["image"]),
    EnsureChannelFirstd(keys=["image"]),
    Orientationd(keys=["image"], axcodes="RAS"),  # standard orientation
    Spacingd(keys=["image"], pixdim=(1.0, 1.0, 1.0), mode="bilinear"),  # resample to standard spacing
])

dataset = Dataset(data=files, transform=transforms)
loader = DataLoader(dataset, batch_size=1)
```
